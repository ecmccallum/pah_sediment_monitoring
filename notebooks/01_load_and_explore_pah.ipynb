```python
# isolate sediment samples only
df_sediment_only = df_sediment[
    df_sediment["libelle_support"].str.contains("sédiment", case=False, na=False)
].copy()

# verify supports
df_sediment_only["libelle_support"].value_counts()
```

```python
df_sediment_only["libelle_fraction"].value_counts()
```

```python
fix_cols = ["libelle_support", "libelle_fraction", "libelle_lieusurv"]

for col in fix_cols:
    if col in df_sediment.columns:
        df_sediment[col] = (
            df_sediment[col]
            .astype(str)
            .str.encode("latin1", errors="ignore")
            .str.decode("utf-8", errors="ignore")
        )
```

```python
df_sediment["libelle_support"].value_counts()
```

```python
df_sediment_only = df_sediment[
    df_sediment["libelle_support"].str.contains("sédiment", case=False, na=False)
].copy()

df_sediment_only["libelle_support"].value_counts()
```

```python
df_sediment_only["libelle_fraction"].value_counts()
```

```python
import numpy as np

df_sediment_only = df_sediment_only.copy()

unit_col = "libelle_unite_resultat"
value_col = "resultat_analyse"

df_sediment_only["result_ug_per_kg_dw"] = np.nan

mask_ugkg = df_sediment_only[unit_col].str.contains(
    "microgramme par kilogramme de matière sèche", na=False
)
df_sediment_only.loc[mask_ugkg, "result_ug_per_kg_dw"] = df_sediment_only.loc[mask_ugkg, value_col]

mask_ngg = df_sediment_only[unit_col].str.contains(
    "nanogramme par gramme de matière sèche", na=False
)
df_sediment_only.loc[mask_ngg, "result_ug_per_kg_dw"] = df_sediment_only.loc[mask_ngg, value_col]

mask_mgkg = df_sediment_only[unit_col].str.contains(
    "milligramme par kilogramme de matière sèche", na=False
)
df_sediment_only.loc[mask_mgkg, "result_ug_per_kg_dw"] = df_sediment_only.loc[mask_mgkg, value_col] * 1000
```

```python
df_sediment_only["result_ug_per_kg_dw"].isna().sum(), len(df_sediment_only)

```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 5))

ax.hist(
    df_sediment_only["result_ug_per_kg_dw"],
    bins=40,
    log=True
)

ax.set_xlabel("Benzo(a)pyrene concentration (µg/kg dry sediment)")
ax.set_ylabel("Number of samples")

ax.set_title("Distribution of benzo(a)pyrene concentrations in coastal sediments")

plt.tight_layout()
plt.show()
```
